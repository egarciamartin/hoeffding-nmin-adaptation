# Baselines plots. 
### Hoeffding Trees with nmin adaptation - Garcia-Martin et al.

- Number of nominal attributes and energy
- Number of numerical attributes and energy
- Number of instances and energy

- Same but for accuracy

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


results = pd.read_csv('results/results_baseline.csv')
results.head(10)
```

```python
# Nominal attributes vs energy

# Energy
vfdt_eng = results.iloc[10:15,6].astype(float)
vfdt_nmin_eng = results.iloc[10:15,4].astype(float)
cvfdt_eng = results.iloc[10:15,5].astype(float)
# Accuracy
vfdt_acc = results.iloc[10:15,3].astype(float)
vfdt_nmin_acc = results.iloc[10:15,1].astype(float)
cvfdt_acc = results.iloc[10:15,2].astype(float)


x_axis = [10,20,30,40,50]
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2,  figsize=(12,9))
# fig.suptitle('Number of nominal attributes vs Accuracy and Energy', y=1.03)

ax1.plot(x_axis,vfdt_eng,  linewidth=1.5, label='VFDT', marker='x', linestyle=':')
ax1.plot(x_axis,vfdt_nmin_eng,linewidth=1.5, label='VFDT-nmin', marker='o',linestyle=':')
ax1.plot(x_axis,cvfdt_eng, linewidth=1.5, label='CVFDT', marker='v', linestyle=':')
ax1.legend(loc='best')
ax1.set_xlabel("Number of Nominal attributes")
ax1.set_ylabel("Total Energy (J)")
ax1.set_xticklabels([ '','10', '', '20', '' , '30', '', '40', '', '50'])

ax3.plot(x_axis,vfdt_eng, linewidth=1.5, label='VFDT', marker='x', linestyle=':')
ax3.plot(x_axis,vfdt_nmin_eng, linewidth=1.5, label='VFDT-nmin', marker='o', linestyle=':')
ax3.legend(loc='best')
ax3.set_xlabel("Number of Nominal attributes")
ax3.set_ylabel("Total Energy (J)")
ax3.set_xticklabels([ '','10', '', '20', '' , '30', '', '40', '', '50'])

ax2.plot(x_axis,vfdt_acc, linewidth=1.5, label='VFDT', marker='x', linestyle=':')
ax2.plot(x_axis,vfdt_nmin_acc, linewidth=1.5, label='VFDT-nmin', marker='o', linestyle=':')
ax2.plot(x_axis,cvfdt_acc, linewidth=1.5, label='CVFDT', marker='v', linestyle=':')
ax2.legend(loc='best')
ax2.set_xlabel("Number of Nominal attributes")
ax2.set_ylabel("Accuracy(%)")
ax2.set_xticklabels([ '','10', '', '20', '' , '30', '', '40', '', '50'])
ax2.set_ylim(0,105)


ax4.plot(x_axis,vfdt_acc, linewidth=1.5, label='VFDT', marker='x', linestyle=':')
ax4.plot(x_axis,vfdt_nmin_acc, linewidth=1.5, label='VFDT-nmin', marker='o', linestyle=':')
ax4.legend(loc='best')
ax4.set_xlabel("Number of Nominal attributes")
ax4.set_ylabel("Accuracy(%)")
ax4.set_xticklabels([ '','10', '', '20', '' , '30', '', '40', '', '50'])
ax4.set_ylim(0,105)




plt.tight_layout()
fig.savefig('figures/eng_acc_nominal.pdf', bbox_inches='tight')

```

```python
# Comparison of energy between nominal and numerical
# Average energy of VFDT, CVFDT, VFDT-nmin
x_axis = [10,20,30,40,50]
nom_vfdt_eng = results.iloc[10:15,6].astype(float)
nom_vfdt_nmin_eng = results.iloc[10:15,4].astype(float)
nom_cvfdt_eng = results.iloc[10:15,5].astype(float)
num_vfdt_eng = results.iloc[5:10,6].astype(float)
num_vfdt_nmin_eng = results.iloc[5:10,4].astype(float)
num_cvfdt_eng = results.iloc[5:10,5].astype(float)


nom1 = pd.concat([nom_vfdt_eng,nom_vfdt_nmin_eng,nom_cvfdt_eng], axis=1)
num1 = pd.concat([num_vfdt_eng,num_vfdt_nmin_eng,num_cvfdt_eng], axis=1)
nom_avg1 = nom1.mean(axis=1)
num_avg1 = num1.mean(axis=1)

nom2 = pd.concat([nom_vfdt_eng,nom_vfdt_nmin_eng], axis=1)
num2 = pd.concat([num_vfdt_eng,num_vfdt_nmin_eng], axis=1)
nom_avg2 = nom2.mean(axis=1)
num_avg2 = num2.mean(axis=1)


fig, (ax1, ax2) = plt.subplots(2, 1,  figsize=(7,7))
# fig.suptitle('Energy consumption of nominal vs numerical attributes', y=1.03)
# f.subplots_adjust(hspace=3.3)


ax1.plot(x_axis,nom_avg1,  linewidth=1.5, label='Nominal (Average VFDT, VFDT-nmin, CVFDT)', marker='x', linestyle=":")
ax1.plot(x_axis,num_avg1,  linewidth=1.5, label='Numerical (Average VFDT, VFDT-nmin, CVFDT)', marker='o', linestyle=":")
ax1.legend(loc='best')
# ax1.set_title("Energy consumption of nominal vs numerical attributes")
ax1.set_xlabel("Number of attributes")
ax1.set_ylabel("Total Energy (J)")
ax1.set_xticklabels([ '','10', '', '20', '' , '30', '', '40', '', '50'])

ax2.plot(x_axis,nom_avg2,  linewidth=1.5, label='Nominal (Average VFDT, VFDT-nmin)', marker='x', linestyle=":")
ax2.plot(x_axis,num_avg2,  linewidth=1.5, label='Numerical (Average VFDT, VFDT-nmin)', marker='o', linestyle=":")
ax2.legend(loc='best')
# ax2.set_title("Energy consumption of nominal vs numerical attributes")
ax2.set_xlabel("Number of attributes")
ax2.set_ylabel("Total Energy (J)")
ax2.set_xticklabels([ '','10', '', '20', '' , '30', '', '40', '', '50'])
plt.tight_layout()
fig.savefig('figures/nom_num.pdf', bbox_inches='tight')
```
