# Real datasets plots. 
### Hoeffding Trees with nmin adaptation - Garcia-Martin et al.

- Each dataset x axis
- Acc and energy



```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib import ticker

results = pd.read_csv('results/results_real.csv')
results_backup = results
display(results.head(10))

# To float
results = results.drop(results.index[0])
results = results.apply(pd.to_numeric,errors='ignore')
# display(results.head(10))
# print(results.dtypes)
```

```python
# Sort based on total energy of VFDT-nmin
results = results.sort_values(by='Total Energy(J)')
display(results.head(10))

# Energy
vfdt_eng = results.iloc[0:6,6].astype(float)
vfdt_nmin_eng = results.iloc[0:6,4].astype(float)
cvfdt_eng = results.iloc[0:6,5].astype(float)

# Accuracy
vfdt_acc = results.iloc[0:6,3].astype(float)
vfdt_nmin_acc = results.iloc[0:6,1].astype(float)
cvfdt_acc = results.iloc[0:6,2].astype(float)

# ticks
ticks = results.iloc[0:6,0]
xticks = ticker.MaxNLocator(4)



x_axis = [0,1,2,3,4,5]
fig, ((ax1, ax2)) = plt.subplots(1, 2,  figsize=(12,5))
# fig.suptitle('Accuracy and energy consumption of real datasets (Sorted by energy of VFDT-nmin)', y=0.95)

ax1.plot(x_axis,vfdt_eng, label='VFDT', marker='x', linestyle=":")
ax1.plot(x_axis,vfdt_nmin_eng, label='VFDT-nmin', marker='o', linestyle=":")
ax1.plot(x_axis,cvfdt_eng, label='CVFDT', marker='v', linestyle=":")
ax1.legend(loc='best')
ax1.set_xlabel("Dataset")
ax1.set_ylabel("Total Energy (J)")
ax1.set_xticks(np.arange(len(ticks)))
ax1.set_xticklabels(ticks)   

  
    
ax2.plot(x_axis,vfdt_acc, label='VFDT', marker='x', linestyle=":")
ax2.plot(x_axis,vfdt_nmin_acc, label='VFDT-nmin', marker='o', linestyle=":")
ax2.plot(x_axis,cvfdt_acc, label='CVFDT', marker='v', linestyle=":")
ax2.legend(loc='best')
ax2.set_xlabel("Dataset")
ax2.set_ylabel("Accuracy(%)")
ax2.set_xticks(np.arange(len(ticks)))
ax2.set_xticklabels(ticks)

fig.savefig('figures/real_sortenergy.pdf', bbox_inches='tight')


```
