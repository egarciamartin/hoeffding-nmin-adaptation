```python
'''
Creating plots for the paper
Plot 1.1 : Acc vs Energy trade-off. Y axis: Total energy. X axis: Acc. 
    Average all datasets for VFDT, VFDT-nmin, and CVFDT
    
Plot 1.2 : Acc vs Energy trade-off. Y axis: Total energy. X axis: Acc. Not averaged

Plot 2: y-axis: total energy. x-axis, VFDT-nmin, VFDT, and CVFDT.
Plot 3: x-axis: datasets. y-axis: energy for each algorithm. VFDT vs VFDT-nmin
Plot 4: X-axis: datasets. y-axis: energy for each algorithm. VFDT vs VFDT-nmin, VFDT-nmin vs CVFDT. in %
Plot 5: X-axis: datasets. y-axis: accuracy for each algorithm. VFDT vs VFDT-nmin, VFDT-nmin vs CVFDT. in %
'''
```

```python
'''
Plot 1.1
'''
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

results = pd.read_csv('../results/results.csv')
results = results.drop([0])

vfdt_acc = results.iloc[:,3].astype(float).mean()
vfdt_nmin_acc = results.iloc[:,1].astype(float).mean()
cvfdt_acc = results.iloc[:,2].astype(float).mean()

vfdt_eng = results.iloc[:,6].astype(float).mean()
vfdt_nmin_eng = results.iloc[:,4].astype(float).mean()
cvfdt_eng = results.iloc[:,5].astype(float).mean()

fig=plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.scatter(vfdt_acc, vfdt_eng, cmap=plt.cm.get_cmap('Set1'), linewidth=3.5, label='VFDT', marker='.', norm=True)
ax.scatter(vfdt_nmin_acc, vfdt_nmin_eng, cmap=plt.cm.get_cmap('Set1'),  linewidth=3.5, label='VFDT-nmin', marker=',',norm=True)
ax.scatter(cvfdt_acc, cvfdt_eng, cmap=plt.cm.get_cmap('Set1'),  linewidth=3.5, label="CVFDT", marker='v',norm=True)
ax.legend(loc='best')
ax.set_yscale('log')
ax.set_title("Accuracy (%) vs Total Energy (J) trade-off")
ax.set_xlabel("Accuracy (%)")
ax.set_ylabel("Total Energy (J)")
ax.set_xlim([0,100.1])

plt.tight_layout()
fig.savefig('../results/figures/acc_eng_avg.pdf', bbox_inches='tight')
```
